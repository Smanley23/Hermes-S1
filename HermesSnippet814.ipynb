```python
import tkinter as tk
from tkinter import ttk, messagebox

root = tk.Tk()
root.title("Hermes â€“ Main Screen")
root.geometry("1000x500")
root.configure(bg='#2c3e50')


distance_map = {
    ("Boston", "New York City"): (215, 4.0),
    ("Boston", "Philadelphia"): (310, 5.5),
    ("Boston", "Cleveland"): (640, 9.5),
    ("Boston", "Baltimore"): (400, 6.5),
    ("Boston", "Washington DC"): (440, 7.0),
    ("New York City", "Boston"): (215, 4.0),
    ("New York City", "Philadelphia"): (95, 2.0),
    ("New York City", "Cleveland"): (460, 7.5),
    ("New York City", "Baltimore"): (190, 3.5),
    ("New York City", "Washington DC"): (225, 4.0),
    ("Philadelphia", "Boston"): (310, 5.5),
    ("Philadelphia", "New York City"): (95, 2.0),
    ("Philadelphia", "Cleveland"): (430, 6.5),
    ("Philadelphia", "Baltimore"): (100, 2.0),
    ("Philadelphia", "Washington DC"): (140, 2.5),
    ("Cleveland", "Boston"): (640, 9.5),
    ("Cleveland", "New York City"): (460, 7.5),
    ("Cleveland", "Philadelphia"): (430, 6.5),
    ("Cleveland", "Baltimore"): (375, 6.0),
    ("Cleveland", "Washington DC"): (375, 6.0),
    ("Baltimore", "Boston"): (400, 6.5),
    ("Baltimore", "New York City"): (190, 3.5),
    ("Baltimore", "Philadelphia"): (100, 2.0),
    ("Baltimore", "Cleveland"): (375, 6.0),
    ("Baltimore", "Washington DC"): (40, 1.0),
    ("Washington DC", "Boston"): (440, 7.0),
    ("Washington DC", "New York City"): (225, 4.0),
    ("Washington DC", "Philadelphia"): (140, 2.5),
    ("Washington DC", "Cleveland"): (375, 6.0),
    ("Washington DC", "Baltimore"): (40, 1.0)
}

flight_map = {
    ("New York City", "Los Angeles"): (2450, 6.0),
    ("New York City", "Austin"): (1750, 4.0),
    ("New York City", "Miami"): (1280, 3.5),
    ("Los Angeles", "Austin"): (1375, 3.5),
    ("Los Angeles", "Miami"): (2340, 5.5),
    ("Austin", "Miami"): (1100, 3.0),
    ("Los Angeles", "New York City"): (2450, 6.0),
    ("Austin", "New York City"): (1750, 4.0),
    ("Miami", "New York City"): (1280, 3.5),
    ("Austin", "Los Angeles"): (1375, 3.5),
    ("Miami", "Los Angeles"): (2340, 5.5),
    ("Miami", "Austin"): (1100, 3.0),
    ("New York City", "Tokyo"): (6750, 14.0),
    ("New York City", "London, England"): (3450, 7.0),
    ("New York City", "Paris, France"): (3625, 7.5),
    ("Tokyo", "New York City"): (6750, 14.0),
    ("London, England", "New York City"): (3450, 7.0),
    ("Paris, France", "New York City"): (3625, 7.5)
}

space_map = {
    "Earth Orbit": (250, 0.5),
    "Moon": (238900, 72.0),
    "Mars": (140000000, 259200.0)
}


main_frame = tk.Frame(root, bg='#2c3e50')
main_frame.pack(fill='both', expand=True)

left_panel = tk.Frame(main_frame, bg='#2c3e50', width=200)
left_panel.pack(side='left', fill='y', padx=(20, 10), pady=20)
left_panel.pack_propagate(False)

content_frame = tk.Frame(main_frame, bg='#34495e', width=500)
content_frame.pack(side='left', fill='both', expand=True, padx=(10, 10), pady=20)
content_frame.pack_propagate(False)

right_panel = tk.Frame(main_frame, bg='#2c3e50', width=280)
right_panel.pack(side='right', fill='y', padx=(10, 20), pady=20)
right_panel.pack_propagate(False)


status_title = tk.Label(right_panel, text="System Status",
                        font=('Ubuntu Mono', 18, 'bold'),
                        bg='#2c3e50', fg='#3498db')
status_title.pack(pady=(20, 10))

status_lines = [
    "AI Core: Online",
    "Navigation: Ready",
    "Sensors: Active",
    "Shield Integrity: Nominal",
    "Life Support: Stable",
    "Comms: Linked",
    "Engine Output: Optimal"
]

for line in status_lines:
    label = tk.Label(right_panel, text=line,
                     font=('Ubuntu Mono', 14),
                     bg='#2c3e50', fg='white')
    label.pack(anchor='w', padx=20, pady=6)


title = tk.Label(left_panel, text="HERMES", font=('Ubuntu Mono', 28, 'bold'),
                 bg='#2c3e50', fg='#3498db')
title.pack(pady=(0, 40))


def show_page(page_name):
    for widget in content_frame.winfo_children():
        widget.destroy()
    
    if page_name == "Hermes":
        label = tk.Label(content_frame, text="Hermes AI Core",
                         font=('Ubuntu Mono', 20, 'bold'), bg='#34495e', fg='white')
        label.pack(pady=(20, 10))

        desc = tk.Label(content_frame, text="Mission AI interface and control center.",
                        font=('Ubuntu Mono', 13), bg='#34495e', fg='#bdc3c7')
        desc.pack(pady=(0, 20))

     
        tk.Label(content_frame, text="Select Transportation:",
                 font=('Ubuntu Mono', 13), bg='#34495e', fg='white').pack(anchor='w', padx=20, pady=5)
        transport_var = tk.StringVar()
        transport_combo = ttk.Combobox(content_frame, textvariable=transport_var, state="readonly",
                                       values=["Land-based travel via car", "Airborne travel via plane", "Extraterrestrial flight"])
        transport_combo.pack(padx=20, pady=5)
        transport_combo.current(0)

        land_cities = ["Boston", "New York City", "Philadelphia", "Cleveland", "Baltimore", "Washington DC"]
        air_cities = ["New York City", "Los Angeles", "Austin", "Miami", "Tokyo", "London, England", "Paris, France"]
        space_destinations = ["Earth Orbit", "Moon", "Mars"]

        tk.Label(content_frame, text="Current Location:", font=('Ubuntu Mono', 13), bg='#34495e', fg='white').pack(anchor='w', padx=20, pady=5)
        current_var = tk.StringVar()
        current_combo = ttk.Combobox(content_frame, textvariable=current_var, state="readonly")
        current_combo.pack(padx=20, pady=5)

        tk.Label(content_frame, text="Destination:", font=('Ubuntu Mono', 13), bg='#34495e', fg='white').pack(anchor='w', padx=20, pady=5)
        dest_var = tk.StringVar()
        dest_combo = ttk.Combobox(content_frame, textvariable=dest_var, state="readonly")
        dest_combo.pack(padx=20, pady=5)

        def update_location_options(event=None):
            choice = transport_var.get()
            if choice == "Land-based travel via car":
                current_combo['values'] = land_cities
                dest_combo['values'] = land_cities
            elif choice == "Airborne travel via plane":
                current_combo['values'] = air_cities
                dest_combo['values'] = air_cities
            elif choice == "Extraterrestrial flight":
                current_combo['values'] = ["Earth"]
                dest_combo['values'] = space_destinations
            current_combo.current(0)
            dest_combo.current(0)

        transport_combo.bind("<<ComboboxSelected>>", update_location_options)
        update_location_options()

        def calculate_travel():
            transport = transport_var.get()
            current = current_var.get()
            dest = dest_var.get()
            result_text = ""
            if transport == "Land-based travel via car":
                route = (current, dest)
                if route in distance_map:
                    dist, time = distance_map[route]
                    result_text = f"Estimated Distance: {dist} miles\nEstimated Travel Time: {time:.1f} hours"
                else:
                    result_text = "Route not supported."
            elif transport == "Airborne travel via plane":
                route = (current, dest)
                if route in flight_map:
                    dist, time = flight_map[route]
                    result_text = f"Estimated Flight Distance: {dist} miles\nEstimated Flight Time: {time:.1f} hours"
                else:
                    result_text = "Flight not available."
            elif transport == "Extraterrestrial flight":
                if dest in space_map:
                    dist, hours = space_map[dest]
                    days = hours / 24.0
                    result_text = f"Distance to {dest}: {dist:,} miles\nEstimated Travel Time: {days:.2f} days"
                else:
                    result_text = "Space destination not supported."
            messagebox.showinfo("Travel Summary", f"Transportation: {transport}\nFrom: {current}\nTo: {dest}\n\n{result_text}")

        tk.Button(content_frame, text="Calculate Travel", font=('Ubuntu Mono', 12),
                  bg='#3498db', fg='white', activebackground='#1abc9c', command=calculate_travel).pack(pady=20)

    else:
        tk.Label(content_frame, text=page_name, font=('Ubuntu Mono', 20, 'bold'),
                 bg='#34495e', fg='white').pack(expand=True, pady=20)

button_labels = [("Hermes", "Hermes"), 
                 ("Diagnostics", "Diagnostics"), 
                 ("Mission Logs", "Mission Logs"), 
                 ("Settings", "Settings")]

for text, page in button_labels:
    b = tk.Label(left_panel, text=text,
                 font=('Ubuntu Mono', 14),
                 bg='#34495e', fg='white',
                 bd=3, relief='raised',
                 padx=10, pady=15,
                 cursor='arrow', width=18)
    b.pack(pady=10)
    b.bind("<Enter>", lambda e, b=b: b.config(bg='#3498db', relief='sunken'))
    b.bind("<Leave>", lambda e, b=b: b.config(bg='#34495e', relief='raised'))
    b.bind("<Button-1>", lambda e, page=page: show_page(page))


show_page("Hermes")

root.mainloop()
```
